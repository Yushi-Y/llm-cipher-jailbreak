```python
import pandas as pd
import string
```

```python
def rot_encoder(text, shift=3):
    """Encodes English text using the Rot cipher with a given number of positions to shift."""
    return ''.join(shift_char(c, shift) for c in text)

def shift_char(c, shift):
    """Shifts a single character by the given amount in the alphabet."""
    if c in string.ascii_lowercase:
        return chr((ord(c) - ord('a') + shift) % 26 + ord('a'))
    elif c in string.ascii_uppercase:
        return chr((ord(c) - ord('A') + shift) % 26 + ord('A'))
    return c  # Non-alphabetic characters remain unchanged



def pig_latin_encoder(text):
    """Encodes English text into Pig Latin."""
    words = text.rstrip(".?!").split()  # Remove punctuation at the end
    transformed_words = [pig_latin_word_encrypt(word) for word in words]
    return " ".join(transformed_words)

def pig_latin_word_encrypt(word):
    """Converts a single word to Pig Latin."""
    vowels = "AEIOUaeiou"
    if word.isdigit():  # If it's a number, still append 'ay'
        return word + "ay"
    if word[0] in vowels:  # Words starting with a vowel
        return word + "ay"
    
    # Find first vowel position
    for i, char in enumerate(word):
        if char in vowels:
            return word[i:] + word[:i] + "ay"
    
    return word + "ay"  # If no vowels, return unchanged with "ay"


def ascii_encoder(text, seven_bit=True):
    """
    Encodes English into ASCII decimal or 7-bit binary representation.
    """
    if seven_bit:
        # Encode to 7-bit binary
        return ' '.join(format(ord(c), '07b') for c in text)
    else:
        # Encode to ASCII decimal
        return ' '.join(str(ord(c)) for c in text)
```

```python
input_file = "/data/kebl6672/llm_jailbreak/harmful_prompts/harmful_prompts.csv"
output_file = "/data/kebl6672/llm_jailbreak/harmful_prompts/ciphered_harmful_prompts.csv"

df = pd.read_csv(input_file)

# Apply encryption functions to harmful prompts
df["harmful_prompt_rot3"] = df["harmful_prompt"].astype(str).apply(lambda x: rot_encoder(x, shift=3))
df["harmful_prompt_rot13"] = df["harmful_prompt"].astype(str).apply(lambda x: rot_encoder(x, shift=13))
df["harmful_prompt_pig_latin"] = df["harmful_prompt"].astype(str).apply(pig_latin_encoder)
df["harmful_prompt_ascii_decimal"] = df["harmful_prompt"].astype(str).apply(lambda x: ascii_encoder(x, seven_bit=False))
df["harmful_prompt_ascii_7bit"] = df["harmful_prompt"].astype(str).apply(lambda x: ascii_encoder(x, seven_bit=True))

# Apply encryption functions to harmless prompts
df["harmless_prompt_rot3"] = df["harmless_prompt"].astype(str).apply(lambda x: rot_encoder(x, shift=3))
df["harmless_prompt_rot13"] = df["harmless_prompt"].astype(str).apply(lambda x: rot_encoder(x, shift=13))
df["harmless_prompt_pig_latin"] = df["harmless_prompt"].astype(str).apply(pig_latin_encoder)
df["harmless_prompt_ascii_decimal"] = df["harmless_prompt"].astype(str).apply(lambda x: ascii_encoder(x, seven_bit=False))
df["harmless_prompt_ascii_7bit"] = df["harmless_prompt"].astype(str).apply(lambda x: ascii_encoder(x, seven_bit=True))

# Save the modified CSV
df.to_csv(output_file, index=False)
print(f"Encrypted CSV saved as {output_file}")
```
